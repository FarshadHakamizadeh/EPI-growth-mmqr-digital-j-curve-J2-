```python
# آمار توصیفی

import pandas as pd
import numpy as np

# بارگذاری  dataset
file_path = 'panel_data_149_EPI_TII_WGI_logGDPpc.xlsx'
df = pd.read_excel(file_path)

# Selecting relevant variables
# Note: Based on previous steps, we need EPI, TII, WGI_mean, HDI, log_GDPpc
variables = ['EPI', 'TII', 'WGI_mean', 'HDI', 'log_GDPpc']

# Descriptive Statistics
desc_stats = df[variables].describe().T[['count', 'mean', 'std', 'min', 'max']]
desc_stats['Skewness'] = df[variables].skew()
desc_stats['Kurtosis'] = df[variables].kurt()

# Display the table for the user to copy
print(desc_stats)
```
