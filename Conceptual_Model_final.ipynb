```python
# مدل مفهومی نهایی


import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, ConnectionPatch

def draw_premium_conceptual_model_fixed():
    fig, ax = plt.subplots(figsize=(16, 8.5))
    ax.set_xlim(0, 17)
    ax.set_ylim(0, 9)
    ax.axis('off')
    
    # --- Colors ---
    c_driver = '#F0F8FF'   
    c_mediator = '#FFF8DC' 
    c_outcome = '#F5F5F5'  
    c_border = '#34495E'
    c_text = '#2C3E50'
    c_arrow_main = '#7F8C8D'
    c_arrow_mod = '#D35400'
    c_arrow_out = '#C0392B'

    def add_box(x, y, width, height, color, label, sublabel):
        # Shadow
        shadow = FancyBboxPatch((x+0.1, y-0.1), width, height, boxstyle="round,pad=0.2",
                                fc='black', alpha=0.08, ec="none")
        ax.add_patch(shadow)
        # Main Box
        box = FancyBboxPatch((x, y), width, height, boxstyle="round,pad=0.2",
                             fc=color, ec=c_border, lw=1.5)
        ax.add_patch(box)
        # Text
        ax.text(x + width/2, y + height*0.65, label, ha='center', va='center',
                fontsize=13, fontweight='bold', color=c_text)
        ax.text(x + width/2, y + height*0.25, sublabel, ha='center', va='center',
                fontsize=16, style='italic', color='#7F8C8D')

    # --- Nodes ---
    add_box(0.5, 6.5, 3.2, 1.2, c_driver, "Technological Infrastructure", "(TII)")
    add_box(0.5, 3.5, 3.2, 1.2, c_driver, "Institutional Quality", "(WGI)")
    add_box(0.5, 0.5, 3.2, 1.2, c_driver, "Human Capital", "(HDI)")
    
    add_box(6.8, 3.5, 3.0, 1.2, c_mediator, "E-Participation", "(EPI)")
    
    add_box(13.2, 3.5, 3.0, 1.2, c_outcome, "Economic Growth", "(Log GDP_pc)")

    # --- Arrows and Text ---
    # استفاده از استایل ظریف تر و تعیین ضخامت خط به جای ضخامت کل بدنه
    arrow_style = "-|>, head_width=0.5, head_length=0.8"
    lw_main = 2.5
    
    # H1
    ax.annotate("", xy=(6.6, 4.5), xytext=(4.0, 6.8),
                arrowprops=dict(arrowstyle=arrow_style, lw=lw_main, color=c_arrow_main))
    ax.text(5.3, 5.9, "H1 (+)", fontsize=12, color=c_border, fontweight='bold', rotation=-36)

    # H3
    ax.annotate("", xy=(6.6, 4.1), xytext=(4.0, 4.1),
                arrowprops=dict(arrowstyle=arrow_style, lw=lw_main, color=c_arrow_main))
    ax.text(5.3, 4.3, "H3 (+)", fontsize=12, color=c_border, fontweight='bold')

    # H2
    ax.annotate("", xy=(6.6, 3.7), xytext=(4.0, 1.4),
                arrowprops=dict(arrowstyle=arrow_style, lw=lw_main, color=c_arrow_main))
    ax.text(5.3, 2.3, "H2 (+)", fontsize=12, color=c_border, fontweight='bold', rotation=36)

    # H4 (Moderation - Curved Arrow)
    mod_arrow = ConnectionPatch(xyA=(2.5, 4.9), xyB=(4.9, 6.0), coordsA="data", coordsB="data",
                                arrowstyle="-|>, head_width=0.5, head_length=0.8", lw=2.5,
                                color=c_arrow_mod, connectionstyle="arc3,rad=-0.2")
    


    
    ax.add_patch(mod_arrow)
    ax.text(2.2, 5.7, "H4: Moderation\n(Synergy vs. Saturation)", 
            fontsize=11, color=c_arrow_mod, fontweight='bold', ha='center', va='bottom')

    # H5 & H6 (Outcome)
    ax.annotate("", xy=(13.0, 4.1), xytext=(10.0, 4.1),
                arrowprops=dict(arrowstyle=arrow_style, lw=lw_main, color=c_arrow_out))

    
    
    ax.text(11.5, 5.5, "H5: Productivity Paradox\n(Digital J-Curve Transition Costs)",
            ha='center', va='bottom', fontsize=11, color=c_arrow_out, fontweight='bold')
    ax.text(11.5, 5.3, "H6: Quantile Heterogeneity\n(Catch-up Effect / Legacy Friction)",
            ha='center', va='top', fontsize=11, color=c_arrow_out, fontweight='bold')

    # --- Title ---
    plt.text(8.5, 8.4, "Conceptual Framework: Drivers, Interaction Effects, and Non-linear Dynamics of E-Participation",
             ha='center', fontsize=15, fontweight='bold', color=c_text)

    plt.tight_layout()
    plt.savefig("Corrected_Conceptual_Model.png", dpi=300, bbox_inches='tight')
    plt.show()

draw_premium_conceptual_model_fixed()
```
