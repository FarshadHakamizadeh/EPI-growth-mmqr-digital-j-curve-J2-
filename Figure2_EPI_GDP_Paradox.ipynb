```python
# بازترسیم شکل جدید 3


import matplotlib.pyplot as plt
import numpy as np

# داده‌های استخراج شده از فصل ۴
quantiles = ['Q10', 'Q25', 'Q50', 'Q75', 'Q90']
coefficients = [-0.062, -0.196, -0.162, -0.299, -0.317]

# مقادیر فرضی برای بازه اطمینان (Confidence Intervals) جهت نمایش علمی‌تر نمودار
ci_upper = [c + 0.05 for c in coefficients]
ci_lower = [c - 0.05 for c in coefficients]

# تنظیمات استایل نمودار
plt.figure(figsize=(9, 6))
plt.style.use('default')

# رسم خط روند و نقاط
plt.plot(quantiles, coefficients, marker='o', markersize=8, color='#b22222', linewidth=2.5, label='EPI Coefficient')

# رسم هاله بازه اطمینان
x_numeric = np.arange(len(quantiles))
plt.fill_between(x_numeric, ci_lower, ci_upper, color='#b22222', alpha=0.15, label='95% Confidence Interval')

# رسم خط مرجع صفر
plt.axhline(y=0, color='black', linestyle='--', linewidth=1.5, alpha=0.7)

# تنظیمات محورها و عناوین
plt.title('Effect of E-Participation (EPI) on Economic Growth (lnGDPpc) Across Quantiles', fontsize=12, fontweight='bold', pad=15)
plt.xlabel('Quantiles', fontsize=11, fontweight='bold')
plt.ylabel('Coefficient of EPI', fontsize=11, fontweight='bold')

# تنظیم محدوده محور Y برای نمایش بهتر مقادیر منفی
plt.ylim(-0.45, 0.1)

# اضافه کردن مقادیر روی نقاط
for i, txt in enumerate(coefficients):
    plt.annotate(f"{txt}", (quantiles[i], coefficients[i]), 
                 textcoords="offset points", xytext=(0,-15), ha='center', fontsize=10, fontweight='bold')

plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(loc='upper right')

# ذخیره نمودار
plt.tight_layout()
plt.savefig('Figure2_EPI_GDP_Paradox.png', dpi=300)
plt.show()
```
